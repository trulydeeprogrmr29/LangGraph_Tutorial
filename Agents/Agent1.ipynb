```python
#our first langgraph implementation
from typing import  Dict,TypedDict
from langgraph.graph import StateGraph  # framework that helps you design and manage the flow of tasks in your application using graph

```

```python
#we now create an AgentState - shared data structure that keeps track of the information
# throughout the execution of the graph
class AgentState(TypedDict):   # our state schema
    message: str

def greeting_node(state: AgentState) -> AgentState:
    """Simple node that adds a greeting message to the state."""
    state["message"] = "Hare Krishna," + state["message"] + "How was your day?"
    return state  
```

```python
graph=StateGraph(AgentState)
graph.add_node("greeter",greeting_node)
graph.set_entry_point("greeter")
graph.set_finish_point("greeter")

app=graph.compile()
```

```python
from IPython.display import Image,display
display(Image(app.get_graph().draw_mermaid_png()))
```

```python
result = app.invoke({"message": " Radha "})
```

```python
result["message"]
```
